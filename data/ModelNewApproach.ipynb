```python
import tensorflow as tf
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv('preprocessed.csv')
data.drop(['ranking', 'Unnamed: 0'], axis = 1, inplace=True)
X, Y = data.drop(['label', 'notlabel'], axis = 1), data[['label', 'notlabel']]
X_train, X_test,y_train, y_test = train_test_split(X, Y, test_size=0.05, shuffle=True)
```

```python
print(len(X_train), len(X_test), len(y_train), len(y_test))
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(256, input_shape=(8,), activation='LeakyReLU', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(128, input_shape=(8,), activation='LeakyReLU'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(64, input_shape=(8,), activation='LeakyReLU'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32, input_shape=(8,), activation='LeakyReLU'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(2, activation="softmax")
])
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=2048)
```

```python
y_pred = model.predict(X_test)
y_pred = np.round(y_pred)
prediction = [y_pred[i][0] for i in range(len(y_pred))]
y_test_final = y_test['label']
print(y_test_final, prediction)
print(metrics.classification_report(y_test_final, prediction))
```

```python
model.evaluate(X_test, y_test, verbose = 1)
```
